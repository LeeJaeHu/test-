```python
users = [
    {"id": 0, "name": "Hero"},
    {"id": 1, "name": "Dunn"},
    {"id": 2, "name": "Sue"},
    {"id": 3, "name": "Chi"},
    {"id": 4, "name": "Thor"},
    {"id": 5, "name": "Clive"},
    {"id": 6, "name": "Hicks"},
    {"id": 7, "name": "Devin"},
    {"id": 8, "name": "Kate"},
    {"id": 9, "name": "Klein"}
]
```

```python
friendships = [(0,1), (0,2), (1,2), (1,3), (2,3), (3,4),
              (4,5), (5,6), (5,7), (6,8), (7,8), (8,9)]
```

```python
for user in users:
    user["friends"] = []
```

```python
users
```

```python
for i, j in friendships:
    #users[i]에서 i는 해당 사용자의 id를 나타내기 때문에
    # 다음과 같이 추가할 수 있다.
    users[i]["friends"].append(users[j])    #j를 i의 친구로 추가
    users[j]["friends"].append(users[i])    # i를 j의 친구로 추가
```

```python
users
```

```python
def number_of_friends(user):
    """user의 친구가 몇 명일까?"""
    return len(user["friends"])    # 각 friends list의 길이
total_connections = sum(number_of_friends(user)
                       for user in users)                  # 24
```

```python
from __future__ import division    # 나눈 값이 정수로 나오면 쓸모가 없다.
num_users = len(users)             # users list의 길이
avg_connections = total_connections / num_users    #2.4
```

```python
# (user_id, number_of_friends)로 구성된 list 생성
num_friends_by_id = [(user["id"], number_of_friends(user))
                    for user in users]

sorted(num_friends_by_id,                                 # 정렬해 보자
      key = lambda (user_id, num_friends): num_friends,    # num_friends 기준으로
      reverse=True)                                       # 제일튼 숫자부터 제일 작은 숫자 순
```
