```python
import numpy as np
```

```python
def Jacobimethod(A,b,x0,e,nmax):
    n = A.shape[0]
    xk = np.zeros(n)
    for k in range(nmax):
        for i in range(n):
            xk[i] = -sum([A[i,j]/A[i,i]*x0[j] for j in range(n) if j != i]) + b[i]/A[i,i]
        print(xk)
        print(x0)
        if sum((xk-x0)**2)**(1/2) <= e:
            print("{}th convergence".format(k))
            return xk,x0
        x0 = xk.copy()
    print(x0,xk,"not convergence")
    return
```

```python
A = np.matrix([[11,2,3],[4,51,6],[7,8,11]])
```

```python
x0=np.array([1,1,1])
```

```python
np.diag(np.diag(A))
```

```python
b = np.ones(3)
```

```python
Jacobimethod(A,b,x0,0.001,100)
```

```python
A.dot(Jacobimethod(A,b,x0,0.001,100)[0])
```
