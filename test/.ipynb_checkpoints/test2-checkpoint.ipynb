```python
lon = ds.variables['lon'][:]
lat = ds.variables['lat'][:]
time = ds.variables['time'][:]
```

```python
tmax_units = ds.variables['time'].units
```

```python
lon_mean = lon.mean()
lat_mean = lat.mean()

m = Basemap(width=5000000, height=3500000,
           resolution='l', projection='stere',\
           lat_ts=40, lat_0=lat_mean, lon_0=lon_mean)
```

```python
lonn, latn = np.meshgrid(lon, lat)
xi, yi = m(lonn, latn)
```

```python
m = Basemap(projection="mill")
```

```python
m.drawcoastlines()
plt.show()
```

```python
m.drawcoastlines()
m.drawcountries()
plt.show()
```

```python
cs = m.pcolor(xi,yi,np.squeeze(time))

m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-100., 181., 10.), labels=[0,0,0,1], fontsize=10)

m.drawcoastlines()
m.drawstates()
m.drawcountries()

cbar = m.colorbar(cs, location='bottom', pad='10%')
cbar.set_label(tmax_units)

plt.title('DJF Maximum Temperature')

plt.show()
```

```python
map = Basemap(projection='mill',llcrnrlon=110., llcrnrlat=20., urcrnrlon=150.,
             urcrnrlat=50., resolution='c')
```

```python
map.drawcoastlines()
map.drawstates()
map.drawcountries()
map.drawlsmask(land_color="Linen", ocean_color='#CCFFFF')
map.drawcountries()

parallels = np.arange(20,50,5.)
meridians = np.arange(110, 150, 5.)
map.drawparallels(parallels, labels=[1,0,0,0],fontsize=10)
map.drawmeridians(meridians, labels=[0,0,0,1],fontsize=10)

lons, lats = np.meshgrid(lon, lat)
x, y = map(lons, lats)

map.pcolormesh(x,y,analysed_sst[0,:,:]-273.15)
map.colorbar(location='right')
```

```python
temp = map.contourf(x,y,analysed_sst[0,:,:])
cb = map.colorbar(temp, "bottom", size="5%", pad="2%")
```
